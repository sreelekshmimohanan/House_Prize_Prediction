```python
import numpy as np
import pandas as pd
import pickle
```

```python
data=pd.read_csv('test.csv')
data
```

```python
file=open("mapping_area_type.txt",'r')
map_area_type=file.read()
file.close()
map_area_type=eval(map_area_type)
map_area_type
```

```python
file=open("mapping_HALL_KITCHEN.txt",'r')
map_HALL_KITCHEN=file.read()
file.close()
map_HALL_KITCHEN=eval(map_HALL_KITCHEN)
map_HALL_KITCHEN
```

```python
file=open("mapping_location_new.txt",'r')
map_location_new=file.read()
file.close()
map_location_new=eval(map_location_new)
dat=list(map_location_new.keys())
print(dat)
```

```python
a=str(dat)
a=a.replace("'",'"')
print(a)
```

```python
file=open("locations.txt","w")
file.write(str(a))
file.close()
```

```python
file=open("mapping_society.txt",'r')
map_society=file.read()
file.close()
map_society=eval(map_society)
type(map_society)
```

```python
data['area_type'].replace(map_area_type,inplace=True)
data['society'].replace(map_society,inplace=True)
data['location_new'].replace(map_location_new,inplace=True)
data['HALL_KITCHEN'].replace(map_HALL_KITCHEN,inplace=True)
data
```

```python
from joblib import load
regressor = load('model.joblib')
```

```python
pred=np.array(regressor.predict(data))
```

```python
out=float(pred[0][0])
print(out)
```

```python
file=open("out.txt",'w')
file.write(str(out))
file.close()
```
